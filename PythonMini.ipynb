## Data Cleaning

```python
#Removes the missing values in the columns
df['Show_Id_num'] = df['Show_Id'].str.extract('(\d+)').astype(int)
df = df.sort_values('Show_Id_num').drop(columns='Show_Id_num')
df.dropna()
```

```python
#fills the missing values in the columns with unknown
df['Show_Id_num'] = df['Show_Id'].str.extract('(\d+)').astype(int)
df = df.sort_values('Show_Id_num').drop(columns='Show_Id_num')
df.fillna('unknown',inplace=True)
```

```python
#removes the specified columns with its values 
df.drop('Description',axis=1)
```

```python
#Counts only the number of movies present in my dataset
df[df['Category'] == 'Movie'].shape[0]
```

```python
#Moves value down by one row
df.shift()
```

```python
#Renames column name with the particular name given to it
df.rename(columns={"Release_Date":"Release_Year"})
```

```python
#Adds a new column from the SAMPLE column already in the dataset
df['Year'] = df['Release_Date'].str[-4:]
```

```python
df.head()
```

```python
#Removes multiple objects after the first comma
df['Country'] = df['Country'].str.split(',').str[0].str.strip()
df['Type'] = df['Type'].str.split(',').str[0].str.strip()
df.head()
```

```python
#Rename a column name permanently for my dataset
df.rename(columns={'Type': 'Genre'}, inplace=True)
print(df)
```

## Data Visualisation

### Number Of Releases By Year

```python
year_counts = df['Year'].value_counts().sort_index()
plt.figure(figsize=(12,6))
plt.bar(year_counts.index, year_counts.values,color='orange',edgecolor='black')
plt.title('Number of Releases By Year', fontweight='bold')
plt.xlabel('Year', fontweight='bold')
plt.ylabel('Count', fontweight='bold')
plt.legend(['Releases'], prop={'weight': 'bold'})
plt.xticks(fontsize=14, fontweight='bold')
plt.yticks(fontsize=15, fontweight='bold')
plt.axis('tight')
plt.grid(True, color="black", linestyle='-', linewidth=2)
plt.show()
```

### Movies Vs TV Shows

```python
category_counts = df['Category'].value_counts().sort_index()
plt.figure(figsize=(8, 8))
plt.pie(
    category_counts.values,             
    labels=category_counts.index,       
    autopct='%1.1f%%',                 
    startangle=90,                       
    colors=['cyan', 'purple'],
    wedgeprops={'edgecolor': 'black','linewidth':2}
)
plt.title('Distribution of Movies vs TV Shows', fontweight='bold')
plt.legend(category_counts.index, title='Category', loc='upper right', prop={'weight': 'bold'})
plt.show()
```

### Genre Of Movies And TV Shows

```python
genre_counts = df['Genre'].value_counts()
heatmap_data=genre_counts.to_frame().T
heatmap_data.index=["Counts"]
plt.figure(figsize=(26,6))
sb.heatmap(
    heatmap_data,
    annot=True,
    annot_kws={"weight": "bold", "fontsize": 12,"color":"black"},
    fmt='d',
    cmap='coolwarm',
    linewidths=3,
    linecolor='black'
)
plt.xticks(fontsize=18, fontweight='bold')
plt.yticks(fontsize=25, fontweight='bold')
plt.xlabel("Genre", fontsize=24, fontweight='bold', color='black')
plt.title('Netflix(Genre): Movies vs TV Shows', fontweight='bold')
plt.show()
```

### Most Reputed Directors

```python
top_directors = df[df['Director'] != "unknown"]['Director'].value_counts().head(20)
plt.figure(figsize=(18,8))
plt.bar(top_directors.index, top_directors.values, color='skyblue', edgecolor='black')
plt.title('Top 20 Directors By Number Of Titles', fontsize=30, fontweight='bold', color='sienna')
plt.xlabel('Directors', fontsize=20, fontweight='bold', color='sienna')
plt.ylabel('Movies or TV Shows Directed', fontsize=18, fontweight='bold', color='sienna')
plt.xticks(rotation=90, fontsize=12, fontweight='bold')
plt.yticks(fontsize=12,fontweight='bold')
plt.grid(True,color='black',linestyle='-',linewidth=1)
plt.show()
```

### Top Countries Producing Netflix Content

```python
country_counts = df['Country'].dropna().value_counts()
top_country = country_counts.head(15)
plt.figure(figsize=(12, 6))
plt.bar(top_country.index, top_country.values, color='pink', edgecolor='black')
plt.title('Top Countries Producing Netflix Content', fontweight='bold')
plt.xlabel('Country', fontweight='bold',fontsize=24,color='indigo')
plt.ylabel('No. Of Movies And TV Shows', fontweight='bold',color='indigo')
plt.xticks(rotation=45,fontsize=12,fontweight='bold',color='sienna')
plt.yticks(fontsize=15,fontweight='bold',color='sienna')
plt.grid(linestyle='-',color='green')
plt.show()
```

### Longest Running Movies And TV Shows

```python
movies = df[df['Category'] == 'Movie'].copy()
movies['Minutes'] = movies['Duration'].str.extract(r'(\d+)').astype(int)
top_movies = movies.sort_values(by='Minutes', ascending=False).head(15)
top_movies.sort_values(by='Minutes', inplace=True)
plt.figure(figsize=(12, 8))
plt.barh(top_movies['Title'], top_movies['Minutes'], color='salmon', edgecolor='black')
plt.xlabel('Duration (Minutes)', fontsize=24,fontweight='bold',color='indigo')
plt.ylabel('Movie Title', fontsize=24,fontweight='bold',color='indigo')
plt.title('Top 15 Longest-Running Movies', fontsize=20, fontweight='bold')
plt.xticks(fontsize=16,fontweight='bold')
plt.yticks(fontsize=16,fontweight='bold')
plt.grid(linestyle='-', alpha=1,linewidth=2,color='purple')
plt.show()
TV_Shows = df[df['Category'] == 'TV Show'].copy()
TV_Shows['Seasons'] = TV_Shows['Duration'].str.extract(r'(\d+)').astype(int)
top_TV_Shows = TV_Shows.sort_values(by='Seasons', ascending=False).head(15)
top_TV_Shows.sort_values(by='Seasons', inplace=True)
plt.figure(figsize=(12, 8))
plt.barh(top_TV_Shows['Title'], top_TV_Shows['Seasons'], color='salmon', edgecolor='black')
plt.xlabel('Duration (Seasons)', fontsize=24,fontweight='bold',color='indigo')
plt.ylabel('TV Show Title', fontsize=24,fontweight='bold',color='indigo')
plt.title('Top 15 Longest-Running TV Shows', fontsize=20, fontweight='bold')
plt.xticks(fontsize=16,fontweight='bold')
plt.yticks(fontsize=16,fontweight='bold')
plt.grid(linestyle='-', alpha=1,linewidth=2,color='purple')
plt.show()
```

# Insights :

##### 1.Releases per year : Netflix’s content production and acquisitions have increased over the years, especially after 2015–2018, showing its rapid global expansion. A dip in recent years, it may be due to incomplete data for the last year or the effect of the pandemic.



#### 2. Movies vs TV shows : Netflix has more movies than TV shows in its dataset. 

#### 3. Genre of movies and TV shows : The most common genres for movies are usually dramas, comedies, and documentaries, while for TV shows they are international series, crime, and kids/family

#### 4. Most reputed directors : A handful of directors contribute a large portion of Netflix’s content . Popular director's repeated collaborations with Netflix show its strategy of building long-term creative partnerships.


#### 5. Top countries producing Netflix content : The U.S. and India are top producers, followed by the U.K. and other international markets.This shows Netflix’s localization strategy, investing in content from different regions to grow its subscriber base globally.



#### 6. Longest-running movies and TV shows : This indicates Netflix mixes short-term binge-worthy content with longer, ongoing franchises.
